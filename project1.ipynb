```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file1='GILD.csv'
file2='IHE.csv'
file3='JNJ.csv'
file4='LLY.csv'
file5='PFE.csv'
file6='SPY.csv'
file7='SYK.csv'
```

```python
gild=pd.read_csv(file1)
ihe=pd.read_csv(file2)
jnj=pd.read_csv(file3)
lly=pd.read_csv(file4)
pfe=pd.read_csv(file5)
spy=pd.read_csv(file6)
syk=pd.read_csv(file7)
```

```python
gild.columns=['Gild Date', 'Gild Open', 'Gild High', 'Gild Low', 'Gild Close', 'Gild Adj Close', 'Gild Volume']
ihe.columns=['IHE Date', 'IHE Open', 'IHE High', 'IHE Low', 'IHE Close', 'IHE Adj Close', 'IHE Volume']
jnj.columns=['JNJ Date', 'JNJ Open', 'JNJ High', 'JNJ Low', 'JNJ Close', 'JNJ Adj Close', 'JNJ Volume']
lly.columns=['LLY Date', 'LLY Open', 'LLY High', 'LLY Low', 'LLY Close', 'LLY Adj Close', 'LLY Volume']
pfe.columns=['PFE Date', 'PFE Open', 'PFE High', 'PFE Low', 'PFE Close', 'PFE Adj Close', 'PFE Volume']
spy.columns=['SPY Date', 'SPY Open', 'SPY High', 'SPY Low', 'SPY Close', 'SPY Adj Close', 'SPY Volume']
syk.columns=['SYK Date', 'SYK Open', 'SYK High', 'SYK Low', 'SYK Close', 'SYK Adj Close', 'SYK Volume']
```

```python
spypct=spy['SPY Close']/279.29*100
gildpct=gild['Gild Close']/70.86*100
plt.plot(spypct)
plt.plot(gildpct)
plt.figure(figsize=(30,30))
spypct.corr(gildpct)
```

```python
spypct=spy['SPY Close']/279.29*100
jnjpct=jnj['JNJ Close']/146.21*100
plt.plot(spypct)
plt.plot(jnjpct)
spypct.corr(jnjpct)
```
